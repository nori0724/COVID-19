```r
library(openxlsx)
```

```r
# 警告メッセージ非表示
options(warn=-1)
```

```r
df <- read.xlsx("../../data/data_covid_fix_name.xlsx")
df <- na.omit(df)
head(df)
```

```r
# 標準化処理
country = df["country"]
date_first = df["date_first"]

df <- as.data.frame(scale(df[,c(2,3,4,5,6,7,8,10,13)]))
df["country"] = country
df["date_first"] = date_first
```

```r
df
```

```r
# データを感染の大小で分ける
early_country <- df[df$date_first < 40, ]
late_country <- df[df$date_first > 40, ]
```

# 重回帰分析

```r
step(model)
```

```r
# fullモデル
model =lm(formula = date_first ~  pop + urb + dist + air + detection, data = df)
summary(model)
AIC(model)
```

- 変数を1つずつ減らしながらp値を確認

### 早期感染国

```r
early_model = lm(formula = date_first ~ hf + pop + urb + gdp + dist + air + detection, data = early_country)
summary(early_model)
AIC(early_model)
```

```r
early_model = lm(formula = date_first ~  pop + urb + gdp + dist + air + detection, data = early_country)
summary(early_model)
AIC(early_model)
```

```r
early_model = lm(formula = date_first ~  pop + gdp + dist + air + detection, data = early_country)
summary(early_model)
AIC(early_model)
```

```r
early_model = lm(formula = date_first ~ gdp + dist + air + detection, data = early_country)
summary(early_model)
AIC(early_model)
```

```r
# 前処理なし
early_model = lm(formula = log(date_first) ~ log(hf) + log(pop) + log(urb) +  log(gdp) +log(dist) + log(air) + log(detection), data = early_country)
summary(early_model)
AIC(early_model)
```
