```r
library(openxlsx)
library(psych)
library(ggplot2)
```

```r
#df = read.csv("../../data/data_covid_fix_name_code_coordinate_ltaly_medianage_hdi_density.csv")
df <- read.xlsx("../../data/data_covid_fix_name.xlsx")
df <- na.omit(df)
df
```

```r
nrow(df)
```

```r
# pop

g <- ggplot(df, aes(x = df$pop))
g <- g + geom_histogram(color="black",fill="light blue", bins = 10)
plot(g)

ggsave("output/pop_hist.eps")
```
