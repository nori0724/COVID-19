### データの分布が歪んでいるときにはclassintを使用する 
- classintについて  
(https://www.rdocumentation.org/packages/rworldmap/versions/1.3-6/topics/mapCountryData)  
- calssintのstyleについて  
(https://www.pu-hiroshima.ac.jp/p/ttetsuji/R/[13]color.html)  
- for numeric data : "pretty", "fixedWidth", "diverging", "logFixedWidth", "quantiles"

```r
library(openxlsx)
library(psych)
library(ggplot2)
library(rworldmap)
library(classInt)
library(RColorBrewer)
```

```r
#df = read.csv("../../data/data_covid_fix_name_code_coordinate_ltaly_medianage_hdi_density.csv")
df <- read.xlsx("../../data/data_covid_fix_name.xlsx")
df <- na.omit(df)
head(df)
```

```r
nrow(df)
```

```r
map_obj <- joinCountryData2Map(df, joinCode="NAME", nameJoinColumn="country")
```

```r
# pop

g <- ggplot(df, aes(x = df$pop))
g <- g + geom_histogram(color="black",fill="light blue", bins = 10) + theme_bw()

plot(g)

ggsave("/Users/masanoritakahashi/Documents/講義/ゼミ/M2/空間生存時間/修士論文/plot/pop_hist.eps")
```

```r
classInt <- classIntervals( df$pop, n=7, style="fisher")
catMethod = classInt$brks

setEPS()
postscript("/Users/masanoritakahashi/Documents/講義/ゼミ/M2/空間生存時間/修士論文/plot/pop_world.eps")
mapCountryData(map_obj, nameColumnToPlot="pop"
               , catMethod=catMethod
               , addLegend = TRUE
              ,mapTitle = "")

dev.off()

mapCountryData(map_obj, nameColumnToPlot="pop"
               , catMethod=catMethod
               , addLegend = TRUE
              ,mapTitle = "")
```

```r
# urb
g <- ggplot(df, aes(x = df$urb))
g <- g + geom_histogram(color="black",fill="light blue", bins = 10) + theme_bw()
plot(g)

ggsave("/Users/masanoritakahashi/Documents/講義/ゼミ/M2/空間生存時間/修士論文/plot/urb_hist.eps")
```

```r
setEPS()
postscript("/Users/masanoritakahashi/Documents/講義/ゼミ/M2/空間生存時間/修士論文/plot/urb_world.eps")
mapCountryData(map_obj, nameColumnToPlot="urb"
               , catMethod="fixedWidth"
               , addLegend = TRUE
              ,mapTitle = "")

dev.off()

mapCountryData(map_obj, nameColumnToPlot="urb"
               , catMethod="fixedWidth"
               , addLegend = TRUE
              ,mapTitle = "")
```
