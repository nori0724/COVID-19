# パラメトリックモデル

### 人間的自由度を用いたモデル  

```r
fs1 <- flexsurvreg(Y ~ log(pf)+log(ef)+log(pop)+log(gdp)+log(urb)+log(dist)+log(air)+log(detection),
                   data = df,dist = "Gompertz")
fs1
```

```r
plot(fs1)
```

### 個人自由度(pf)、経済的自由度(ef)を用いたモデル

```r
fs2 <- flexsurvreg(Y ~ log(hf)+log(pop)+log(gdp)+log(urb)+log(dist)+log(air)+log(detection),
                   data = df,dist = "Gompertz")
fs2
```

```r
plot(fs2, col="blue",)
```

# Cox比例ハザードモデル  

### 人間的自由度を用いたモデル¶

```r
# 変数選択
stepAIC(coxph(Y ~ log(hf) + log(pop)+log(gdp)+log(urb)+log(dist)+log(air)+log(detection),data = df))
```

```r
# 比例ハザードモデル
cox1 <- coxph(formula = Y ~ log(hf) + log(pop) + log(gdp) + log(urb) +  log(dist) + log(detection), data = df)
cox1

# AIC
AIC(cox1)
```

```r
# 残差plot
scatter.smooth(residuals(cox1, type="deviance"))
abline(h=0,lty=3,col=2)
```

```r
# 比例ハザード性の確認
cox1.zph<- cox.zph(cox1)
cox1.zph
```

## 人間的自由度を用いたモデル(距離で層別化)

```r
hist(df$dist)
```

```r
# 中央値の算出
dist_median = median(df$dist)

# 中央値以上のものは1、以下は0
df$dist_median = "0"
df[df$dist > dist_median, ][,"dist_median"]="1"
```

```r
# 変数選択
stepAIC(coxph(Y ~ log(hf) + log(pop)+log(gdp)+log(urb)+log(air)+log(detection)+ strata(dist_median),data = df))
```

```r
cox1_strata <- coxph(formula = Y ~ log(hf) + log(pop) + log(gdp) + log(urb) +  log(air) + strata(dist_median), data = df)
cox1_strata

AIC(cox1_strata)
```

```r
# 残差plot
scatter.smooth(residuals(cox1_strata, type="deviance"))
abline(h=0,lty=3,col=2)
```

```r
# 比例ハザード性の確認
cox1_strata.zph<- cox.zph(cox1_strata)
cox1_strata.zph
```

```r
# シェーンフィールド残差plot
par(mfrow=c(3,2))
plot(cox1_strata.zph,df=2)
```

```r
# 残差plot
scatter.smooth(residuals(cox1_strata, type="schoenfeld"))
abline(h=0,lty=3,col=2)
```
