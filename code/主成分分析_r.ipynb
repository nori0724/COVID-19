```r
library(openxlsx)
```

```r
df <- read.xlsx("../data/data_covid_fix_name.xlsx")
```

```r
df <- na.omit(df)
# index reset
rownames(df) <- 1:nrow(df)
head(df)
```

```r
# data = df[,c(2,3,4,5,7,8,9,10,13)]
data = df[,c(2,3,4,5,6,9,10,13)]
head(data)
```

```r
data["pop"] = log(data["pop"])
data["air"] = log(data["air"])
data["gdp"] = log(data["gdp"])

# 標準化
data_scale = scale(data)
```

# 主成分分析  

#### 引数 “scale” について  
- 関数 “prcomp” に、引数 “scale=T” を指定  
⇒ 相関行列から主成分分析を行う  
- 関数 “prcomp” に、引数 “scale=F” を指定   
⇒ 分散共分散行列から主成分分析を行う  

#### 分析結果に関数 “summary” を適用  
- Standard deviation (標準偏差)  
- Proportion of Variance (寄与率)  
- Cumulative Proportion (累積寄与率)  

```r
result <- prcomp(data_scale, scale=T)
summary(result)

# 第3主成分までで78%を説明
```

## 固有ベクトルの出力

```r
round(result$rotation, 3)
```

## 因子負荷量  
 プログラムの概略    
- １行目 ： 因子負荷量の計算    
固有ベクトル（result$rotation）と、対応した固有値の平方根（result$sdev）との積をとる  
- sweep 関数の使い方は、apply 関数とよく似ている  
参考URL : R-Tips 24節 「applyファミリー」  
http://cse.naro.affrc.go.jp/takezawa/r-tips/r/24.html  

### 目的　　
- 主成分に強く寄与している変数を見つけることが重要
- 因子負荷量が１か-1に近い因子ほど、主成分に強く寄与している  
- 因子負荷量をプロットすることにより、 主成分に寄与している因子を視覚的に捉えることができる  

```r
fc.l <- sweep(result$rotation, MARGIN=2, result$sdev, FUN="*")
# col <- c("Pop", "urb", "gdp", "dist", "pf", "ef", "detection", "air")
col <- c("Pop", "urb", "gdp", "dist", "hf", "Date_first", "detection", "air")
plot(fc.l[,1], pch=col, ylim=c(-1,1), main="PC1")
plot(fc.l[,2], pch=col, ylim=c(-1,1), main="PC2")
plot(fc.l[,3], pch=col, ylim=c(-1,1), main="PC3")
plot(fc.l[,4], pch=col, ylim=c(-1,1), main="PC4")
```

### 因子負荷量の解釈(2次元)  

- 第１主成分に強く寄与するが、第２主成分にはほとんど寄与しない因子  
urb, pf, gdp, ef, dist  
- 第２主成分に強く寄与する因子  
Pop

```r
plot(fc.l[,1], fc.l[,2], pch=col, xlim=c(-1,1), ylim=c(-1,1), main="PC1, PC2")
```
