## クラスターごとの可視化

```r
g4<-ggplot(df,aes(x=date_first)) #x軸はSepal.Length。ヒストグラムのy軸は度数なので、ここでは指定不要
g4<-g4+geom_histogram(aes(fill=df$cluster), #種類ごとに処理して塗り分け
                      #bins=30, #棒は３０本立てる
                      size=10, #横に表示するレジェンドのマーク?の大きさ
                      　alpha=0.5) #透過性
g4
```

```r
overlap_hist <- function(df, x, factor) {
   g5 <- ggplot(df,  
            aes(x = x,　
                y = ..density..,  
                colour = factor,
                fill = factor 
                )
            )
    g5 <- g5 + geom_histogram(   # ヒストグラム
        position = "identity",
        alpha = 0.5
        )

    g5 <- g5 + geom_density(     # 密度曲線
        stat = "density",
        position = "identity",
        alpha = 0.5
        )


    #描画
    return(g5)
}
```

```r
# date_first
overlap_hist(df, df$date_first, df$cluster)
```

```r
# pop
overlap_hist(df, log(df$pop), df$cluster)
```

```r
# urb
overlap_hist(df, df$urb, df$cluster)
```

```r
# gdp
overlap_hist(df, log(df$gdp), df$cluster)
```

```r
# dist
overlap_hist(df, df$dist, df$cluster)
```

```r
# hf
overlap_hist(df, df$hf, df$cluster)
```

```r
# ef
overlap_hist(df, df$ef, df$cluster)
```

```r
# pf
overlap_hist(df, df$pf, df$cluster)
```

```r
# detection
overlap_hist(df, df$detection, df$cluster)
```

```r
# air
overlap_hist(df, log(df$air), df$cluster)
```

# 生存時間解析 

```r
#生存時間object
Y = Surv(df$date_first, df$status ==1)
Y1 = Surv(df1$date_first, df1$status ==1)
Y2 = Surv(df2$date_first, df2$status ==1)
Y3 = Surv(df3$date_first, df3$status ==1)
Y4 = Surv(df4$date_first, df4$status ==1)
```

```r
# KM
par(family="Osaka")
kmfit1 = survfit(Y1 ~ 1)
plot(kmfit1,  mark.t=TRUE, xlab="time", ylab = "survival probabilities", main="クラスター1の生存曲線")

kmfit2 = survfit(Y2 ~ 1)
plot(kmfit2,  mark.t=TRUE, xlab="time", ylab = "survival probabilities", main="クラスター2の生存曲線")

kmfit3 = survfit(Y3 ~ 1)
plot(kmfit3,  mark.t=TRUE, xlab="time", ylab = "survival probabilities", main="クラスター3の生存曲線")

kmfit4 = survfit(Y4 ~ 1)
plot(kmfit4,  mark.t=TRUE, xlab="time", ylab = "survival probabilities", main="クラスター4の生存曲線")
```

# Cox比例ハザードモデル  

```r
stepAIC(coxph(Y1 ~ log(hf) + log(pop) + log(gdp) + log(urb) + log(dist) + log(air) + log(detection) ,data = df1))
```

```r
#cox_cluster1 <- coxph(Y1 ~ hf+pop+gdp+urb+dist+air+detection,data = df1)
cox_cluster1 <- coxph(formula = Y1 ~ log(hf) + log(pop) + log(gdp) + log(urb) + log(dist), data = df1)
cox_cluster1
```

```r
AIC(cox_cluster1)
```

```r
# 比例ハザード性の確認(シェーンフィールド残差)
# 仮説が棄却されると比例ハザードの仮定が充たされていない可能性があることを示唆する。
zph_cluster1<- cox.zph(cox_cluster1)
zph_cluster1
```

```r
scatter.smooth(residuals(cox_cluster1, type="deviance"))
abline(h=0,lty=3,col=2)
```

# 層別Cox比例ハザードモデル¶

```r
#生存時間object
Y = Surv(df$date_first, df$status ==1)
```

```r
stepAIC(coxph(Y ~ log(hf) + log(pop) + log(gdp) + log(urb) + log(dist) + log(air) + log(detection) + strata(cluster) ,data = df))
```

```r
cox_strata <- coxph(formula = Y ~ log(hf) + log(pop) + log(gdp) + log(urb) + log(dist) + log(detection) + strata(cluster), data = df)
cox_strata

AIC(cox_strata)
```

```r
# 残差plot
scatter.smooth(residuals(cox_strata, type="deviance"))
abline(h=0,lty=3,col=2)
```

```r
# 比例ハザード性の確認
cox_strata.zph<- cox.zph(cox_strata)
cox_strata.zph
```

```r
# シェーンフィールド残差plot
par(mfrow=c(3,2))
plot(cox_strata.zph,df=2)
```

```r
# 残差plot
scatter.smooth(residuals(cox_strata, type="schoenfeld"))
abline(h=0,lty=3,col=2)
```
