# 新規感染者数がmaxの行を抽出

```python
df.groupby("countriesAndTerritories")["cases"].max().reset_index()
```

```python
ddf = df.groupby("countriesAndTerritories")

# 新規感染者数がmaxのindex抽出
df_cases_max = df.loc[ddf['cases'].idxmax(),:]

# 欠損があるとエラー
#df_drop=df.dropna()
df_drop = df[df["Cumulative_number_for_14_days_of_COVID-19_cases_per_100000"].isnull()==False]
ddf = df_drop.groupby("countriesAndTerritories")
# 14日間の累積感染者数がmaxのindex抽出
df_cumulative_max = df_drop.loc[ddf['Cumulative_number_for_14_days_of_COVID-19_cases_per_100000'].idxmax(),:]
```

# 収束までにかかった時間(移動平均ver)  

```python
# fix_countryをcountry に変更
# mergeするため
df_cumulative_max.columns = ['dateRep', 'day', 'month', 'year', 'cases', 'deaths',
       'countriesAndTerritories', 'geoId', 'countryterritoryCode',
       'popData2019', 'continentExp',
       'Cumulative_number_for_14_days_of_COVID-19_cases_per_100000',
       'country', 'datetime']
```

```python
# 必要なところだけ
df_cumulative_max = df_cumulative_max[["datetime", "country", "cases", "Cumulative_number_for_14_days_of_COVID-19_cases_per_100000"]]
```

```python
# 名寄せ
df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("United States of America", "USA")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("United Republic of Tanzania", "Tanzania")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("United Kingdom", "UK")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("Trinidad and Tobago", "Tobago")

#df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("Timor-Leste", "Timor Leste")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("Trinidad and Tobago", "Tobago")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("North Macedonia", "Macedonia")
```

```python
# 名寄せ
df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("Brunei Darussalam", "Brunei")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("Cote dIvoire", "Ivory Coast")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("Czechia", "Czech Republic")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("Guinea Bissau", "Guinea-Bissau")

df_cumulative_max["country"] = df_cumulative_max["country"].str.replace("Timor Leste", "Timor-Leste")
```

```python
origin_df = pd.read_excel("/Users/masanoritakahashi/Documents/講義/ゼミ/M2/空間生存時間/論文/covid19/data_covid_fix_name.xlsx")
```

```python
print(len(origin_df))
origin_df = pd.merge(origin_df, df_cumulative_max, on="country", how="left")
```

```python
origin_df["china_start"]="2019-12-30"
origin_df["china_start"] = pd.to_datetime(origin_df["china_start"])

# 各国でコロナ初感染の日にち計算
origin_df["start_date"] = origin_df["china_start"] + origin_df["date_first"].apply(lambda x:datetime.timedelta(days=x))
# コロナ感染から収束までの日数計算
origin_df["corona_converge_days"] = (origin_df["datetime"] - origin_df["start_date"])
origin_df["corona_converge_days"] = origin_df["corona_converge_days"].dt.days
```

```python
# 相関行列のヒートマップを描く
data=origin_df[['pop', 'urb', 'gdp', 'dist', 'hf', 'pf', 'ef', 'date_first', 'detection', 'air', 'corona_converge_days']]
corr = data.corr()
plt.figure(figsize=(16, 9))
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns,
            yticklabels=corr.columns)
plt.savefig("correlation_heatmap.png")
```
